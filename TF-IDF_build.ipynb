```python
import pandas as pd
import numpy as np
import nltk

from nltk import word_tokenize
```

```python
df = pd.read_csv('bbc_text_cls.csv')
```

```python
df.head()
```

```python
# populate word2idx
# convert document into sequences of ints /ids / indices

idx = 0
word2idx = {}
tokenized_doc = []
for doc in df['text']:
    tokens = word_tokenize(doc.lower())
    doc_as_int = []
    for token in tokens:
        if token not in word2idx:
            word2idx[token] = idx # key = token, value = index
            idx += 1

        doc_as_int.append(word2idx[token])
    tokenized_doc.append(doc_as_int)
```

```python
word2idx 
```

```python
type(word2idx)
```

Reverse mapping

```python
idx2word = {v:k for k, v in word2idx.items()}
idx2word
```

```python
# stored with list
idx2word_list = [k for k in word2idx.keys()]
idx2word_list
```

```python
# number of documents
N = len(df['text'])
N
```

```python
V = len(word2idx)
V
```

```python
# instantiate term-frequency matrix
# could also use count vetorizer

tf = np.zeros((N,V))

# populate term frequency counts
for i, doc_as_int in enumerate(tokenized_doc):
    for j in doc_as_int:
        tf[i,j]+=1
```

```python
tf
```

```python
type(tf)
```

```python
# Compute IDF
document_count = np.sum(tf>0,axis=0)  # document frequency (shape=(V,))  axis = 0 -> sum over each row(word)
idf = np.log(N/ document_count) # N = documents, document_count = term appears documents
```

```python
idf
```

```python
# Compute TF-IDF
tf_idf = tf*idf
```

```python
# 固定 random 的結果
np.random.seed(1)
```

```python
# pick a random document, show the top 5 terms
i = np.random.choice(N)
row = df.iloc[i]
print('Label :', row['labels'])
print('Text :', row['text'].split('\n', 1)[0])
print('Top 5 terms:')
scores = tf_idf[i]
indices = (-scores).argsort()

for j in indices[:5]:
    print(idx2word[j])
    print(int(tf[i][j]))
```
